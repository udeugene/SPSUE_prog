```python
import numpy as np
import pandas as pd
from decimal import Decimal
import matplotlib.pyplot as plt
%matplotlib inline
import math
from matplotlib import style
from scipy.misc import derivative
import inspect

from IPython.display import display, clear_output

style.use('ggplot')
style.use('seaborn-ticks')
```

```python
from search import *
```

```python
def f(x):
    return x**4 + 4*(x**3) + 7*(x**2) + 2*x + 1

def fib(n, cache={0: 0, 1: 1}):
    if n not in cache:
        cache[n] = fib(n - 1, cache) + fib(n - 2, cache)
    return cache[n]
```

```python
from search import *
l = np.arange(1, 7, dtype=np.float32)

rep = pd.DataFrame(columns = ['e', 'method', 'steps',"x*", 'f(x*)'])
for e in l:
    model2 = Passive_search(f, -3, 2, 10**-e, prec = e.astype(int))
    model2.data()
    passive = [f'10e-{int(e)}', 'Passive_search', len(model2.dfdata), round(model2.res()[0], e.astype(int)), round(model2.res()[1], e.astype(int))]
    rep.loc[len(rep)] = passive
    
    model1 = Golden_search(f, -3, 2, 10**-e, prec = e.astype(int))
    model1.fit()
    gold = [f'10e-{int(e)}', 'Golden_search', len(model1.data), min(round(model1.data.iloc[:, 2][len(model1.data) - 1], e.astype(int)), round(model1.data.iloc[:, 2][len(model1.data) - 1], e.astype(int))), min(round(model1.data.iloc[:, 3][len(model1.data) - 1], e.astype(int)), round(model1.data.iloc[:, 3][len(model1.data) - 1], e.astype(int)))]
    rep.loc[len(rep)] = gold

    model3 = Fibonacci_search(f, -3, 2, 10**-e, prec = e.astype(int))
    model3.fit()
    fib = [f'10e-{int(e)}', 'Fibonacci_search', len(model3.data), round(model3.res[0], e.astype(int)), round(model3.res[1], e.astype(int))]
    rep.loc[len(rep)] = fib
    
    model4 = Secant_search(f, -3, 2, 10**-e, prec = e.astype(int))
    model4.fit()
    secant = [f'10e-{int(e)}', 'Secant_search', len(model4.data), round(model4.res[0], e.astype(int)), round(model4.res[1], e.astype(int))]
    rep.loc[len(rep)] = secant
    
    model5 = Newton_search(f, -3, 2, 10**-e, prec = e.astype(int))
    model5.fit()
    newton = [f'10e-{int(e)}', 'Newton_search', len(model5.data), round(model5.res[0], e.astype(int)), round(model5.res[1], e.astype(int))]
    rep.loc[len(rep)] = newton
```

```python
rep.set_index(['e', 'method'], inplace=True)
```
